```python
def Plot_xH():
    LineWidth = 2
    FontSize = 18
    h5f = h5py.File(Template_Simulation_File, 'r')
    z = h5f['global/z'][:]
    xH = h5f['global/xH'][:]
    xH_C2Ray = h5f['global/xH_C2Ray'][:]
    h5f.close()

    fig, ax = plt.subplots()
    ax.grid(True, which='both', linewidth = 0.3)  # `which='both'` enables major and minor grids
    fig.set_size_inches(8, 5)
    plt.plot(z, xH, 'k', linewidth=LineWidth, label = '21cmFAST')
    plt.plot(z, xH_C2Ray, 'r', linewidth=LineWidth, label = 'C2Ray')
    plt.xlabel('$z$',fontsize=FontSize,fontname='Times New Roman')
    plt.ylabel('$x_{\mathrm{H}}$',fontsize=FontSize,fontname='Times New Roman')
    plt.xticks(size=FontSize)
    plt.yticks(size=FontSize)
    plt.legend(fontsize=FontSize,loc = 'lower right')
    plt.xlim([5, 20])
    plt.ylim([-0.1, 1.2])
    plt.tight_layout()
    plt.show()

Plot_xH()
```
